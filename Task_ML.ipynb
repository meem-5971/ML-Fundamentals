```python
import pandas as pd
import numpy as np
```

```python
df =pd.read_csv("/content/drive/MyDrive/Dataset-20240827T051357Z-001/Dataset/train.csv")
df
```

Dropping Unneccessary columns

```python
train_1=df.copy()
```

```python
train_1=train_1.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
train_1
```

Finding the categories of string type data

```python
pd.unique(train_1['Sex'])
```

```python
pd.unique(df['Cabin'])
```

```python
pd.unique(train_1['Embarked'])
```

```python
pd.unique(train_1['Age'])
```

Finding the missing values

```python
df.isnull().sum()
```

Replacing the missing values of "Embarked" with Mode

```python
train_1["Embarked"].fillna(df['Embarked'].mode()[0],inplace=True)
train_1.isnull().sum()
```

Replacing the missing value of Age with Median

```python
train_1["Age"].fillna(df['Age'].median(),inplace=True)
train_1.isnull().sum()
```

Encoding the string data

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
train_1["Sex"]=lb.fit_transform(train_1["Sex"])
train_1
```

```python
from sklearn.preprocessing import OneHotEncoder
oh = OneHotEncoder(sparse_output=False, handle_unknown='error',drop='first')

```

```python
dfn=oh.fit_transform(train_1[["Embarked"]])
dfn
```

```python
ohendef=pd.DataFrame(dfn)
ohendef
```

```python
train_1=train_1.drop(["Embarked"],axis=1)
train_1
```

```python
train_1=train_1.join(ohendef)
train_1
```

Feature Selection

```python
x=train_1.drop('Survived',axis=1)
x
```

```python
y=train_1['Survived']
y
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
x.columns = x.columns.astype(str)
fit_features=SelectKBest(score_func=f_classif)
fit_features.fit(x,y)
```

```python
fs= pd.DataFrame(fit_features.scores_,index=x.columns,columns=['score value'])
fs
```

```python
sp=fs.nlargest(8,'score value')
sp
```

#Model Fitting

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lrg=LinearRegression()
```

```python
lrg.fit(x_train,y_train)

lrg.score(x_train,y_train)*100
```

```python
lrg.score(x_test,y_test)*100
```

```python
y_pred_lrg=lrg.predict(x_test)
y_pred_lrg
```
